```python
f = open('./LPF.txt')
lines = f.readlines()
header = lines[1]
lines = lines[2:]
f.close()
```

```python
con = np.zeros((len(lines), 3), dtype='float32')
for idx, line in enumerate(lines):
    ele = lines[idx].split('       ')
    if '' in ele:
        ele.remove('')
    for i, e in enumerate(ele):
        if e == '' or e == ' ':
            i -= 1
            continue
        e = e.replace(' ', '')
        e = e.replace('\n', '')
        if e != '':
            x = float(e)
            con[idx][i] = float(x)
```

```python
header = header.replace('\n', '').split('\t')
```

```python
df = pd.DataFrame(con, columns=header)
df.to_csv('./LPF.csv', index=False)
```
